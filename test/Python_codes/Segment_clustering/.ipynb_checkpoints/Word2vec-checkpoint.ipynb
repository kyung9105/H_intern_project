```python
data = pd.read_csv('santa-fe_review.csv')
```

```python
word_data = data['text'].values.tolist()
```

```python
word_data[0].split(' ')
```

```python
# corpus generaion
sentances = []
for row in word_data:
    # words = nonLetterTokenizer(str(row['review']))
    words = str(row).split()
    # words.append( row['title'] )
    sentances.append( words )
```

```python
model = gensim.models.Word2Vec(sentances, min_count=1, size=100,window=2)
```

```python
model_index = model.wv.index2word
model_index
```

```python
model_vector = pd.DataFrame(model[model_index], index=model_index)
```

```python
model_vector
```

```python
from scipy.spatial import distance
from tqdm import tqdm
```

```python
words = model_vector.index
func = distance.euclidean
sim_matrix = []
for word in tqdm(words):
    similarites = []
    word_vec = model_vector.loc[word].values.tolist()
    for word2 in words:
        word2_vec = model_vector.loc[word2].values.tolist()
        similarites.append(round(func(word_vec, word2_vec),5))

    sim_matrix.append(similarites)
```

```python
from sklearn.preprocessing import normalize
```

```python
model_vector.index
```

```python
normal_sim = pd.DataFrame((normalize(sim_matrix)),index=model_vector.index, columns=model_vector.index)
```

```python
normal_sim
```

```python
voca = normal_sim.columns.values
```

```python
test = [ str(x) for x in data['text'].values.tolist() ] 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(vocabulary=voca)
context = test
X = vectorizer.fit_transform(context)
TDM_matrix = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names())
```

```python
columns = ['price','mpg','design']
```

```python
ext_sim_matrix = normal_sim[columns]
```

```python
test = pd.DataFrame(np.dot(TDM_matrix, ext_sim_matrix), columns=columns)
```
