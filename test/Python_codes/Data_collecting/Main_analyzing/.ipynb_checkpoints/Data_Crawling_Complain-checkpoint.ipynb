# links 불러오는 함수 작성
* **3단계의 link**를 불러온다

```python
def get_post_link(url): # url은 hyundai/santafe/2017 즉, make,model,year을 변경해줘야한다. 
    target_url = url 
    res = urllib.request.urlopen(target_url)

    html = res.read()
    soup = BeautifulSoup(html, 'html.parser')

    root_path = soup.find('div', id='graph').find('ul')
    second_list = root_path.find_all('li')

    third_path =[] 
    # 최종 path를 담은 list

    for path in second_list:
        second_path = target_url + path.find('a').get('href')
        res = urllib.request.urlopen(second_path)
        html = res.read()
        soup = BeautifulSoup(html, 'html.parser')
        third_list = soup.find('div', id='graph').find('ul').find_all('li')

        for path in third_list:
            third_path.append(second_path + path.find('a').get('href'))
            # 문자열인 path을 연결해서 최종 third_list path를 만들어낸다.
            
    return third_path
```

```python
def get_post_review(link): # 반복문을 통해 모든 link를 받아들인다.
    
    Big_category = link.split('/')[-2]
    Small_category = link.split('/')[-1].split('.')[0]
    # category는 link에 포함되어 있는 내용이다.
    
    url = link
    res = urllib.request.urlopen(url)
    html = res.read()
    soup = BeautifulSoup(html, 'html.parser')

    
    complaint_path = soup.find('div', id='pcomments').find_all('div', class_='complaint')
    # 각 link당 homepage에 보여지는 각각의 'Complain 항목' 들
    
    review_head =  [comment_path.find('div', class_='cheader') 
                    for comment_path in complaint_path ]
    
    # 각각의 'Complain'마다 date / year / title / sub_title 을 포함하는 구역 
    
    B_category = [Big_category for x in range(len(review_head)) ]
    S_category = [Small_category for x in range(len(review_head)) ]
    
    date = [ head.find('div', class_='pdate').get_text('-').split('-')[0] 
            for head in review_head ]
    year = [ head.find('div', class_='pdate').get_text('-').split('-')[1]
            for head in review_head ]
    title = [ head.find('h3').text
            for head in review_head ]
    sub_title = [ head.find('ul').get_text('-')
            for head in review_head ]  

    
    review_body = [comment_path.find('div', class_='comments')
                    for comment_path in complaint_path ]
    
    # 각각의 'Complain'마다 text / city / country 을 포함하는 구역 
    
   
    try:
        text =  [ body.find('div', itemprop='reviewBody').get_text(strip=True) for body in review_body ]
    except AttributeError:
        text = []
        for body in review_body:
            if body.find('div', class_='ad'):
                test = body.find('script').get_text(strip=True)
                text.append(body.get_text(strip=True).split(test)[1])
            else:
                text.append(body.get_text(strip=True))
    
    # text는 html의 구성이 지멋대로 이므로 일반적인 상황(가장 빈도가 높은 상황)을 try처리하고 나머지는 예외처리
    # error가 발생하는 부분을 밑에 Test에서 확인해가면서 최적의 함수를 작성
        
    city =  [ body.find('p', class_='userinfo').get_text(strip=True).split()[1].replace(',','') 
    for body in review_body ]
    country =  [ body.find('p', class_='userinfo').get_text(strip=True).split()[-1]
    for body in review_body ]
    
    
    return{
        'title' : title,
        'sub_title' : sub_title,
        'year' : year,
        'date' : date,
        'country' : country,
        'city' : city,
        'text' : text,
        'B_category' : B_category,
        'S_category' : S_category
    }
```

```python
zip_code= [('hyundai','Santa_fe', 2017),('hyundai','Santa_fe', 2016),('hyundai','Santa_fe', 2015),('hyundai','Santa_fe', 2014),
           ('hyundai','Santa_fe',2013),('hyundai','Santa_fe',2012)]
```

```python
model_year=2017

for (make, model) in zip_code:
    try:
        links = get_post_link('http://www.carcomplaints.com/'+make+'/'+model+'/'+str(model_year)+'/')
    except urllib.request.HTTPError as e:
            print('hi')
```

```python
# Make/Model/year 을 변경해주면된다.

year_list = [2017,2016,2015,2014,2013,2012]

for year in year_list:
    model_year=year
    for make, model in tqdm(zip_code):
        try:
            links = get_post_link('http://www.carcomplaints.com/'+make+'/'+model+'/'+str(model_year)+'/')
        except (urllib.request.HTTPError, AttributeError) as e:
                continue

        else:
            for link in tqdm(links):
                try:
                    get_post_review(link)
                except AttributeError:
                    continue
                else:    
                    review_table = pd.DataFrame(get_post_review(link))
                    review_table['make_name'] = make
                    review_table['model_name'] = model
                    review_table['model_year'] = model_year

                    result_table = pd.concat([review_table , pre_table])

                    pre_table = result_table
```

```python
for make, model,model_year in tqdm(zip_code):
        links = get_post_link('http://www.carcomplaints.com/'+make+'/'+model+'/'+str(model_year)+'/')
        for link in tqdm(links):
            review_table = pd.DataFrame(get_post_review(link))
            review_table['make_name'] = make
            review_table['model_name'] = model
            review_table['model_year'] = model_year

            result_table = pd.concat([review_table , pre_table])

            pre_table = result_table
```
