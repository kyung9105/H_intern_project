```python
hap_suv_review2_17 = hap_suv_review2
hap_suv_review2_17
```

```python
comfort = pd.read_json(hap_suv_review2_17['comfortRatingDto'].to_json(), orient='index')
interior = pd.read_json(hap_suv_review2_17['interiorRatingDto'].to_json(), orient='index')
perform = pd.read_json(hap_suv_review2_17['performanceRatingDto'].to_json(), orient='index')
reliab = pd.read_json(hap_suv_review2_17['reliabilityRatingDto'].to_json(), orient='index')
safety = pd.read_json(hap_suv_review2_17['safetyRatingDto'].to_json(), orient='index')
tech = pd.read_json(hap_suv_review2_17['technologyRatingDto'].to_json(), orient='index')
value = pd.read_json(hap_suv_review2_17['valueRatingDto'].to_json(), orient='index')
dum_table = hap_suv_review2_17[['make_name','model_name']]

hap_table = pd.concat([dum_table, comfort, interior, perform, reliab, safety, tech, value],axis=1)
hap_table = hap_table.dropna()
hap_table.shape
```

```python
hap_table.head(3)
```

```python
column_list = ['frontSeats',
       'gettingInOut', 'noiseAndVibration', 'rearSeats', 'rideComfort',
       'cargoStorage', 'instrumentation', 'interiorDesign','logicOfControls', 'qualityOfMaterials', 'acceleration', 'braking',
              'roadHolding', 'shifting', 'steering',
       'dealershipSupport', 'electronics', 'engine','repairFrequency', 'transmission', 'activeSafety', 'headlights',
       'outwardVisibility', 'parkingAids', 'rainSnowTraction','bluetooth', 'climateControl', 'entertainment', 'navigation',
              'usbPorts', 'fuelEconomy', 'maintenanceCost',
       'purchaseCost', 'resaleValue', 'warranty']

for column in column_list:
    hap_table.loc[hap_table[column]=='EXCELLENT',column] = float(5)
    hap_table.loc[hap_table[column]=='OKAY',column] = float(3)
    hap_table.loc[hap_table[column]=='TERRIBLE',column] = float(1)

    
    
for column in column_list:
    hap_table[column] = hap_table[column].astype(float)
```

## 평점 순위 살펴보기

```python
save_table = hap_table.groupby('model_name').mean()
```

```python
## Maxcruz 42개의 세부 평점 중 순위가 1등인 것 -> 좋은 점수를 받은 항목
## Maxcruz 42개의 세부 평점 중 순위가 3등 or 4등인 것 -> 나쁜 점수를 받은 항목
## 2개의 그룹으로 나누고 관찰

save_table_rank = save_table.T.rank(ascending=False)
index1 = save_table_rank.T[save_table_rank.T['santa-fe']==4].index
index2 = save_table_rank.T[save_table_rank.T['santa-fe']==3].index
index3 = save_table_rank.T[save_table_rank.T['santa-fe']==1].index
```

```python
first = save_table.loc[index2.values]
second = save_table.loc[index1.values]

pros_maxcruz = pd.concat([first, second])
pros_maxcruz
```
