```python
os.chdir('C:\\Users\\kb910\\Google 드라이브\\경배끄10\\인턴\\csv파일\\spec\\Segment관련_SUV')
```

```python
data = pd.read_csv('test5.csv')
```

```python
data = data.drop(['Unnamed: 0'], axis=1)
```

```python
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import train_test_split
```

```python
train_data = data[data['pass_vol'].notnull()]
```

```python
feature = train_data.columns
print(feature)
```

```python
feautre_x = ['length', 'width', 'height',
       'cargo_space','base_weight']

## select
feautre_y = 'pass_vol'

X = train_data[feautre_x].values
y = train_data[feautre_y].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
                                                    
model = RandomForestRegressor(n_estimators=1000, n_jobs=5, random_state=1)
model.fit(X_train, y_train)
```

```python
## select
weight_list = data[data['pass_vol'].isnull()].index
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

score_train = r2_score(y_train, y_train_pred)
score_test = r2_score(y_test, y_test_pred)

print("R2 - Train : %.3f, Test : %.3f" %(score_train, score_test))
```

## 회귀에 적용한 값을 다시 table에 옮기기

```python
test = []
for index in weight_list:
    pred_feature  = data.loc[index][[18,20,21,22,25]].values
    test.append(float(model.predict(pred_feature)))
    
data.loc[data['pass_vol'].isnull(), 'pass_vol'] = test
```
