### Gathering the Dataset

```python
import requests
import json
import time
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA
from tqdm import tqdm
```

```python
# Set your header according to the form below
# :: (by /u/)

# Add your username below
hdr = {'User-Agent': 'windows:r/politics.single.result:v1.0' +
       '(by /u/)'}
url = 'https://www.reddit.com/r/politics/.json'
req = requests.get(url, headers=hdr)
json_data = json.loads(req.text)

data_all = json_data['data']['children']
num_of_posts = 0
for _ in tqdm(range(10)): # 실습환경에 따라 간단하게 10개 만 실행
    time.sleep(2) # robot 에 따름
    
    last = data_all[-1]['data']['name']
    url = 'https://www.reddit.com/r/politics/.json?after=' + str(last) #가장 최근 수집한 정보에 따른 다음 기사 수집
    
    req = requests.get(url, headers=hdr)
    data = json.loads(req.text)
    data_all += data['data']['children']
    
    if num_of_posts == len(data_all):
        break
    else:
        num_of_posts = len(data_all)
```

### Labeling our Data

```python
## Example

sia = SIA()
post= data_all[0]['data']['title']
print(post)
print("This Sentiment Score is"+ str(sia.polarity_scores(post)))
```

```python
## pos_news_titles.txt 긍정 단어사전
## neg_news_titles.txt 부정 단어사전

sia = SIA()
pos_list = []
neg_list = []
for post in data_all:
    res = sia.polarity_scores(post['data']['title'])

    print(res)
    
    if res['compound'] > 0.2:
        pos_list.append(post['data']['title'])
    elif res['compound'] < -0.2:
        neg_list.append(post['data']['title'])

with open("pos_news_titles.txt", "w", encoding='utf-8',
          errors='ignore') as f_pos:
    for post in pos_list:
        f_pos.write(post + "\n")

with open("neg_news_titles.txt", "w", encoding='utf-8',
          errors='ignore') as f_neg:
    for post in neg_list:
        f_neg.write(post + "\n")
```

#### neg : 부정 , neu : 중립 , pos : 긍정 , compound : (-1 ~ +1)의 감정분석 지표

# Sentiment Analyzing2
## NLTK 내장 모듈(VADER)로 감정분석 하기
* **VADER(Valence Aware Dictionary and Sentiment Reasoner)** Process
    1. Open file(lexicon.txt) : sid.lexicon (단어별 score가 기록되어 있다.)
    2. Filtering 'a','i',.... : 한 글자로 이루어진 단어는 제외시키는 부분이다. nltk로 불용어처리를 하면 보다 정확하고 편리한데 여기서는 length을 체크하여 1보다 작거나 같으면 필터링하게끔 되어있다.
 
    3. Check punctuation and count : 구두점이나 느낌표, 물음표 등등의 글자들이 있는지를 검사한다.

    4. sentiment score를 계산해주는 핵심적인 부분이다. 총 3부분으로 되어있다.

        4.1 Check negative : aren't나 cannot 같은 부정적 동사를 체크하는 단계이다. 부정적 동사가 있으면 -0.74의 가중치를 곱한다.   
        4.2 Check capital : 대문자는 강조의 의미로 보고 0.733을 positive일 경우 더하고 negative일 경우 뺀다.   
        4.3 Weight booster words : amazingly, barely와 같은 단어들은 수식어로 명사에 가중치를 부여한다. 긍정일 경우 0.293, 부정일 경우 -0.293을더한다. 추가로, 수식어의 위치에 따라 다른 가중치가 부여된다. 단어의 앞의 앞 단어라면 0.9가 곱해지고 단어 앞의 수식어는 0.95의 가중치가 곱해진다. 즉, 단어와 가까운 수식어가 더 높은 점수를 받는다.   
        
    5. 'But' is reverse score : But의 경우 내용의 반전이기 때문에 단어의 앞에 있을 경우 0.5를 곱하고 뒤에있을 경우 1.5를 곱한다.   
           ex) Okay, but 일경우 *0.5  but i'm okay, 일 경우 *1.5

    6. Weight emphasis : 느낌표의 경우 강조의 의미이므로 느낌표 하나당 0.295점을 주고 최대 4개까지 점수를 부여한 후 단어 점수에 더한다.

    7. Normalization score : -1~1 사이로 normalization 하는 단계이다. compound 계산에 사용된다. compound는 expect max value이다.

    8. Calculate proportion : 7까지 계산된 sentiment score를 비율로 나타낸다. positive score / ( pos score + neg score + neu score)처럼 계산함.
    

### VADER 의 특징

## 모델 Testing & Example

```python
sentences = ["VADER is smart, handsome, and funny.", # positive sentence example
  "VADER is smart, handsome, and funny!", # punctuation emphasis handled correctly (sentiment intensity adjusted)
  "VADER is very smart, handsome, and funny.",  # booster words handled correctly (sentiment intensity adjusted)
  "VADER is VERY SMART, handsome, and FUNNY.",  # emphasis for ALLCAPS handled
  "VADER is VERY SMART, handsome, and FUNNY!!!",# combination of signals - VADER appropriately adjusts intensity
  "VADER is VERY SMART, really handsome, and INCREDIBLY FUNNY!!!",# booster words & punctuation make this close to ceiling for score
  "The book was good.",         # positive sentence
  "The book was kind of good.", # qualified positive sentence is handled correctly (intensity adjusted)
  "The plot was good, but the characters are uncompelling and the dialog is not great.", # mixed negation sentence
  "A really bad, horrible book.",       # negative sentence with booster words
  "At least it isn't a horrible book.", # negated negative sentence with contraction
   ":) and :D",     # emoticons handled
   "",              # an empty string is correctly handled
  "Today sux",     #  negative slang handled
  "Today sux!",    #  negative slang with punctuation emphasis handled
  "Today SUX!",    #  negative slang with capitalization emphasis
  "Today kinda sux! But I'll get by, lol" # mixed sentiment example with slang and constrastive conjunction "but"
]
```

```python
sid = SIA()
for sentence in sentences:
    print(sentence)
    ss = sid.polarity_scores(sentence)
    for k in sorted(ss):
        print(' {0}: {1}, '.format(k, ss[k]), end='')
    print()
```

```python
neg_sentence_list=[]
pos_sentence_list=[]
neutral_list=[]

for sentence in last_sentences:

    ss = sid.polarity_scores(sentence)
    if(ss['compound'] <0):
        neg_sentence_list.append(sentence)
    elif(ss['compound'] >0):
        pos_sentence_list.append(sentence)
    else:
        neutral_list.append(sentence)
```

## Step3 : 긍정, 부정리뷰의 언급단어 비교

```python
neg_sentence_list
```

```python
tokenizer = RegexpTokenizer(r'\w+')
stop_words = set(stopwords.words('english'))

all_words_neg = []
all_words_neg2 = []
    
for line in neg_sentence_list:
    words = tokenizer.tokenize(line)
    for w in words:
        if w.lower() not in stop_words:
            all_words_neg.append(w.lower())
            
            
tagged_list = nltk.pos_tag(all_words_neg)
for tag_word in tagged_list:
    if(tag_word[1] == 'NN' or tag_word[1] == 'JJ' or tag_word[1] == 'JJS'):
        all_words_neg2.append(tag_word)
    
neg_res = nltk.FreqDist(all_words_neg2)

print(neg_res.most_common())
```

```python
neg_res.plot(10)
```

```python
tokenizer = RegexpTokenizer(r'\w+')
stop_words = set(stopwords.words('english'))

all_words_pos = []
all_words_pos2 = []
    
for line in pos_sentence_list:
    words = tokenizer.tokenize(line)
    for w in words:
        if w.lower() not in stop_words:
            all_words_pos.append(w.lower())
            
tagged_list = nltk.pos_tag(all_words_pos)
for tag_word in tagged_list:
    if(tag_word[1] == 'NN' or tag_word[1] == 'JJ' or tag_word[1] == 'JJS'):
        all_words_pos2.append(tag_word)
    
pos_res = nltk.FreqDist(all_words_pos2)

print(pos_res.most_common())
```
