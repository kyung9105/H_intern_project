# 상관분석 진행

```python

suv_table = pd.read_csv('last_csv_sell_spec.csv')
print(suv_table.shape)
suv_table.head(5)
suv_table = suv_table.drop(['Unnamed: 0'], axis=1)
```

```python
suv_table.columns
```

```python
cor_table = suv_table[['horsepower','MPG_avg','baseMSRP','Y2016']]
```

```python
corr = cor_table.corr()
```

```python
corr
```

```python
sns.heatmap(corr)
```

```python
cor_table['baseMSRP'].describe()
```

```python
cor_table['baseMSRP'][0]
```

```python
low = []
mid_50 = []
mid_75 = []
high = []

for index in range(len(cor_table)):
    if cor_table['baseMSRP'][index] >= 26308 and cor_table['baseMSRP'][index] <= 39340:
        low.append(index)
    elif cor_table['baseMSRP'][index] > 39340 and cor_table['baseMSRP'][index] <= 50412:
        mid_50.append(index)
    elif cor_table['baseMSRP'][index] > 50412 and cor_table['baseMSRP'][index] <= 62797:
        mid_75.append(index)
    else:
        high.append(index)
        
low_suv = cor_table.loc[low]
mid_50_suv = cor_table.loc[mid_50]
mid_75_suv = cor_table.loc[mid_75]
high_suv = cor_table.loc[high]
```

```python
from scipy.stats.stats import pearsonr
```

```python
pearsonr(low_suv['horsepower'], low_suv['Y2016'])
```

```python
fig, ((ax1,ax2),(ax3,ax4))  = plt.subplots(2,2)
fig.set_size_inches(18,10)
sns.regplot(x='horsepower' ,y='Y2016', data=low_suv, ax=ax1)
sns.regplot(x='horsepower' ,y='Y2016', data=mid_50_suv, ax=ax2)
sns.regplot(x='horsepower' ,y='Y2016', data=mid_75_suv, ax=ax3)
sns.regplot(x='horsepower' ,y='Y2016', data=high_suv, ax=ax4)
```

```python
fig, ((ax1,ax2),(ax3,ax4))  = plt.subplots(2,2)
fig.set_size_inches(18,10)
sns.regplot(x='MPG_avg' ,y='Y2016', data=low_suv, ax=ax1)
sns.regplot(x='MPG_avg' ,y='Y2016', data=mid_50_suv, ax=ax2)
sns.regplot(x='MPG_avg' ,y='Y2016', data=mid_75_suv, ax=ax3)
sns.regplot(x='MPG_avg' ,y='Y2016', data=high_suv, ax=ax4)
```
