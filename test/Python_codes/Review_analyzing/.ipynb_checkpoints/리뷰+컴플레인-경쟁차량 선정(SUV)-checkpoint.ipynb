# Maxcruz 각 경쟁차량 비중

```python
def Maxcruz_to_SUV(table):
    
    voca_list = ['highlander','durango','cx-9','pilot','pathfinder','4runner','flex','traverse','acadia','explorer','murano','cherokee']
    
    for voca in voca_list:

        test = [ str(x) for x in table.values.tolist() ] 

        from sklearn.feature_extraction.text import CountVectorizer
        vectorizer = CountVectorizer(vocabulary=[voca])
        context = test
        X = vectorizer.fit_transform(context)
        TDM_matrix = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names())

        for column in TDM_matrix.columns:
            TDM_matrix.loc[TDM_matrix[column] != 0, column] = 1

        TDM_matrix.loc[('sum_count')] = [ TDM_matrix[x].sum() for x in TDM_matrix.columns ]

        score = round(TDM_matrix.loc[('sum_count')].sum() / len(table),3)*100

        print("Maxcruz의" +str(voca)+ "경쟁 비중 : "+str(score)+"%")
```

```python
SUV15 = data[data['year_dummy']==2015]
SUV16 = data[data['year_dummy']==2016]
SUV17 = data[data['year_dummy']==2017]
```

## SUV15년도

```python
def review_count(year):
    voca = ['tacoma','ridgeline','colorado','canyon','frontier']
    test = [ str(x) for x in year['text'].values.tolist() ] 

    from sklearn.feature_extraction.text import CountVectorizer
    vectorizer = CountVectorizer(vocabulary=voca)
    context = test
    X = vectorizer.fit_transform(context)
    TDM_matrix = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names())

    for column in TDM_matrix.columns:
        TDM_matrix.loc[TDM_matrix[column] != 0, column] = 1

    TDM_matrix.loc[('sum_count')] = [ TDM_matrix[x].sum() for x in TDM_matrix.columns ]

    return TDM_matrix.loc[('sum_count')].sum() / len(year)
```

```python
print(review_count(SUV15), len(SUV15))
print(review_count(SUV16), len(SUV16))
print(review_count(SUV17), len(SUV17))
```

## 각 경쟁SUV에서 Maxcruz비중

```python
os.chdir('C:\\Users\\kb910\\Google 드라이브\\경배끄10\\인턴\\csv파일\\리뷰들\\SUV리뷰')
data2 = open('all_suv_review(15~17).json').read()
data2 = pd.DataFrame(json.loads(data2))
all_suv_review = data2
```

```python
SUV17 = all_suv_review[all_suv_review['year_dummy']==2017]
SUV16 = all_suv_review[all_suv_review['year_dummy']==2016]
```

```python
SUV_table = pd.concat([SUV16, SUV17])
```

```python
model_list = all_suv_review['model_name'].unique()
model_list
```

```python
all_suv_review[all_suv_review['model_name']=='murano']
```

```python
def SUV_to_Maxcruz(table):
    
    model_list = all_suv_review['model_name'].unique()
    
    for model in model_list:
        table = all_suv_review[all_suv_review['model_name']==model]
        voca = ['santa-fe','santa','fe']
        test = [ str(x) for x in table['text'].values.tolist() ] 

        from sklearn.feature_extraction.text import CountVectorizer
        vectorizer = CountVectorizer(vocabulary=voca)
        context = test
        X = vectorizer.fit_transform(context)
        TDM_matrix = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names())

        for column in TDM_matrix.columns:
            TDM_matrix.loc[TDM_matrix[column] != 0, column] = 1

        TDM_matrix.loc[('sum_count')] = [ TDM_matrix[x].sum() for x in TDM_matrix.columns ]

        score = round(TDM_matrix.loc[('sum_count')].sum() / len(table)*100,3)
        
        print(str(model)+"의 Maxcruz 경쟁 비중 : "+str(score)+"%"+'  / 리뷰 counting :  '+str(TDM_matrix.loc[('sum_count')].sum()) +  '   / 전체리뷰개수 :'+str(len(table)))
        print('\n')
```

```python
SUV_to_Maxcruz(SUV_table)
```
