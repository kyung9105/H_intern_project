```python
data = open('real_final_suv_seg.json').read()
data = pd.DataFrame(json.loads(data))
```

```python
data.index = [ x for x in range(len(data))]
data.columns
```

```python
from sklearn.cluster import KMeans
import mglearn
```

```python
data['interior'] = data[['pass_capa', 'f_head', 'f_leg','cargo_space','pass_vol',
       'f_shoulder', 'f_hip', 's_head', 's_leg', 's_shoulder', 's_hip']].sum(axis=1)

data['exterior'] = data[['wheelbase', 'length', 'width', 'height',
       'track_width_front', 'track_width_rear']].sum(axis=1)
```

```python
feature = ['pass_capa', 'f_head', 'f_leg','f_shoulder', 'f_hip', 's_head', 's_leg', 's_shoulder', 's_hip',
        'wheelbase', 'length', 'width', 'height','track_width_front', 'track_width_rear',
           'interior','exterior','cargo_space','pass_vol','base_weight']
cluster_value = data[feature].values
```

```python
kmeans = KMeans(n_clusters=3, random_state=1)
kmeans.fit(cluster_value)
```

```python
labels = ['Largesize_SUV','Midsize_SUV','MidLarge_SUV' ]
plot_x = pd.DataFrame(cluster_value)[15]
plot_y = pd.DataFrame(cluster_value)[16]

fig, ax = plt.subplots(1, figsize=(15,8))
mglearn.discrete_scatter(cluster_value[:, 15], cluster_value[:, 16], kmeans.labels_, markers='o')


for i, txt in enumerate(data['model_name'].values):
    ax.annotate(txt, (cluster_value[i][15],cluster_value[i][16]+1.6), color='black', 
                size=14, ha='center')

plt.xlabel('Interior Size', fontsize=15)
plt.ylabel('Exterior Size', fontsize=15)
plt.tick_params(axis='both', which='major', labelsize=15)
ax.legend(labels,loc='upper center',bbox_to_anchor=(0.5, -0.08),
          fancybox=True, shadow=True, ncol=3, prop={'size':15})

ax.annotate('MaxCruz', xy=(cluster_value[31][15], cluster_value[31][16]), xytext=(cluster_value[1][15]+20, cluster_value[31][16]-10),
            arrowprops=dict(facecolor='black', shrink=0.05), fontsize=20, color='red')

ax.annotate('Santa-fe', xy=(cluster_value[32][15], cluster_value[32][16]), xytext=(cluster_value[32][15]+20, cluster_value[32][16]-10),
            arrowprops=dict(facecolor='black', shrink=0.02), fontsize=20, color='red')

plt.plot([plot_x.min(),plot_x.max()],[plot_y.min(),plot_y.max()],color='red')
plt.savefig('첫번째 Segment')
plt.title('첫번째 Segment(소비자 구매인식)', fontsize=15, weight='bold')
plt.show()
```

## 두번째 Segment

```python
data = open('real_final_suv_seg2.json').read()
data = pd.DataFrame(json.loads(data))
```

```python
suv_table = data
```

```python
SUV_D_cluster1 = suv_table[suv_table["Segment_y"] == 'Mid_SUV']
SUV_D_cluster2 = suv_table[suv_table["Segment_y"] == 'MidLarge_SUV']
SUV_D_cluster = pd.concat([SUV_D_cluster1, SUV_D_cluster2])
```

```python
D_price_hp = SUV_D_cluster[['baseMSRP','horsepower']].values
SUV_D_cluster.index = SUV_D_cluster['model_name']
```

```python
def elbow(x):
    sse = []
    for i in range(1,11):
        km=KMeans(n_clusters=i, init='k-means++', random_state=0)
        km.fit(x)
        sse.append(km.inertia_)
        
    plt.plot(range(1,11), sse, marker='o')
    plt.xlabel('클러스터 개수')
    plt.ylabel('SSE')
    plt.show

elbow(D_price_hp)
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(D_price_hp)
```

```python
SUV_D_cluster.index = [x for x in range(len(SUV_D_cluster))]
SUV_D_cluster[SUV_D_cluster['model_name']=='santa-fe-sport'].index
```

```python
labels = ['1','2,','3','4','5']
plot_x = pd.DataFrame(D_price_hp)[0]
plot_y = pd.DataFrame(D_price_hp)[1]

fig, ax = plt.subplots(1, figsize=(15,10))
mglearn.discrete_scatter(D_price_hp[:, 0], D_price_hp[:, 1], kmeans.labels_, markers='o')


for i, txt in enumerate(SUV_D_cluster['model_name']):
    ax.annotate(txt, (D_price_hp[i][0]+100,D_price_hp[i][1]+3), color='black', 
                size=14, ha='center')

plt.xlabel('Price', fontsize=15)
plt.ylabel('HorsePower', fontsize=15)
plt.tick_params(axis='both', which='major', labelsize=15)
ax.legend(labels,loc='upper center', bbox_to_anchor=(0.5, -0.07),
          fancybox=True, shadow=True, ncol=6, prop={'size':15})

ax.annotate('MaxCruz', xy=(D_price_hp[11][0], D_price_hp[11][1]), xytext=(D_price_hp[11][0]+10, D_price_hp[11][1]+50),
            arrowprops=dict(facecolor='black', shrink=0.05), fontsize=20, color='red')

ax.annotate('Santa-fe', xy=(D_price_hp[12][0], D_price_hp[12][1]), xytext=(D_price_hp[12][0]+1000, D_price_hp[12][1]-50),
            arrowprops=dict(facecolor='black', shrink=0.05), fontsize=20, color='red')

plt.plot([plot_x.min(),plot_x.max()],[plot_y.min(),plot_y.max()])
plt.title('두번째 세그먼트(가격)', fontsize=15, weight='bold')
plt.savefig('두번째 세그먼트(가격)')
plt.show()
```

```python
kmeans.labels_
# 0은 Seg1
# 4는 Seg3
# 3은 Seg5
# 2는 Seg2
# 1은 Seg4
```

```python
segment_suv2 = pd.concat( [ pd.DataFrame(kmeans.labels_), pd.DataFrame(SUV_D_cluster['model_name'])] ,axis=1)
```

```python
segment_suv2.columns = ['Segment','model_name']
segment_suv2.loc[segment_suv2['Segment']==0, 'Segment'] = 'Seg1'
segment_suv2.loc[segment_suv2['Segment']==1, 'Segment'] = 'Seg4'
segment_suv2.loc[segment_suv2['Segment']==2, 'Segment'] = 'Seg2'
segment_suv2.loc[segment_suv2['Segment']==3, 'Segment'] = 'Seg5'
segment_suv2.loc[segment_suv2['Segment']==4, 'Segment'] = 'Seg3'
```

```python
test = pd.merge(SUV_D_cluster, segment_suv2, on='model_name')
```

```python
test.to_csv('real_final_suv_seg3.csv')
test.to_json('real_final_suv_seg3.json')
```
